```python
# Import
import sys
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Add the src directory to the Python path
module_path = os.path.abspath(os.path.join('..', 'src'))

if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
# Import modules
from causalml import dgp
```

```python
df = dgp.generate_grouped_data(base_trend_slope=0.02, parallel_trends=True)

grouped_data = df.groupby(['time', 'treated_unit'])['Y_it'].mean().reset_index()

# Plotting the data
plt.figure(figsize=(10, 6))
sns.lineplot(data=grouped_data, x='time', y='Y_it', hue='treated_unit', marker='o')
plt.title('Grouped Average $Y_{it}$ by Treatment Status Over Time')
plt.xlabel('Time')
plt.ylabel('Average $Y_{it}$')
plt.grid(True)
plt.show()
```
